```python
#Series
import pandas as pd
x = [3,4,5,6,7,8]
var = pd.Series(x)
print(var)
print(type(var))
print(var[2])
```

```python
y = [4,5,6,8]
y_var = pd.Series(y, index=['a', 's', 'd', 'f'], dtype='float', name='python')
print(y_var)

```

```python
dic = {"name":['python','c','c++','java'], "por":[12,13,14,15], "rank":[1,4,3,2]}
var1 = pd.Series(dic)
print(var1)
```

```python
s = pd.Series(12, index=[1,2,3,4,5,6,7])
print(s)
print(type(s))
```

```python
s1 = pd.Series(12, index=[1,2,3,4,5,6,7])
s2 = pd.Series(12, index=[1,2,3,4])
print(s1+s2)
```

```python
#DataFrame
import pandas as pd
l = [1,2,3,4,5,6]
df = pd.DataFrame(l)
print(type(df))
print(df)
```

```python
d = {"a":[1,2,3,4,5],"b":[1,2,3,4,5],"c":[1,2,3,4,5]}
df1 = pd.DataFrame(d,columns=["a","b"])
df2 = pd.DataFrame(d,columns=["a","b"], index=["a","s","d","f","g"])
print(type(df1))
print(df1)
print(type(df2))
print(df2)
```

```python
d = {"a":[1,2,3,4,5],"b":[1,2,3,4,5],"c":[1,2,3,4,5]}
dataframe = pd.DataFrame(d)
print(dataframe)
print(dataframe["c"][4])
```

```python
list_1 = [[1,3,5,6],[2,4,6,8]]
var_2 = pd.DataFrame(list_1)
print(type(var_2))
print(var_2)
```

```python
sr = {"d":pd.Series([1,2,3,4]),"r":pd.Series([5,6,7,8])}
var_3 = pd.DataFrame(sr)
print(type(var_3))
print(var_3)
```

```python
#Arithmetic-Operation
import pandas as pd
x = pd.DataFrame({"A":[5,6,7,8,9],"B":[2,4,6,8,10]},index=['a','s','d','f','g'])
print(x)
```

```python
x["C"] = x["A"] + x["B"]
x
```

```python
x["C"] = x["A"] - x["B"]
x
```

```python
x["C"] = x["A"] * x["B"]
x
```

```python
x["C"] = x["A"] / x["B"]
x
```

```python
x1 = pd.DataFrame({"A":[10,20,30,40],"B":[15,25,35,45]})
x1
```
